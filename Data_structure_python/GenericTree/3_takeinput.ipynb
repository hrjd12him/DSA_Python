```python
class treenode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right  =None

def treeinput():
    rootdata = int(input())
    if rootdata == -1:
        return
    rootnode = node(rootdata)
    rootleft = treeinput()
    rootright = treeinput()

    rootnode.left = rootleft
    rootnode.right = rootright

    return rootnode


def printtree(root):
    if root == None:
        return 
    print(root.data,end=':')
    if root.left != None:
        print(root.left.data,end=',')
    if root.right != None:
        print(root.right.data)
    print()
    printtree(root.left)
    printtree(root.right)

root = treeinput()
printtree(root)
    
```
